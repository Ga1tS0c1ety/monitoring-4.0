```python
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio
from scipy.fft import fft, fftfreq
from scipy.signal import hilbert
from IPython.display import display, Markdown

# Configurer le renderer Plotly pour utiliser "browser" par d√©faut
pio.renderers.default = "browser"

# Fr√©quence d'√©chantillonnage
fs = 25600  # Hz
delta_t = 1 / fs  # Intervalle de temps entre deux √©chantillons (en secondes)

# Nombre d'√©chantillons pour chaque fichier
N_sains = len(df_sains)
N_outer_race = len(df_outer_race)

# G√©n√©rer l'axe temporel pour les deux fichiers
time_sains = np.arange(0, N_sains * delta_t, delta_t)
time_outer_race = np.arange(0, N_outer_race * delta_t, delta_t)

# G√©n√©rer les fr√©quences pour l'axe des abscisses (pour la FFT)
frequencies_sains = fftfreq(N_sains, 1/fs)
frequencies_outer_race = fftfreq(N_outer_race, 1/fs)

# Afficher les dur√©es totales
display(Markdown("""
## Informations sur les signaux ‚è±Ô∏è

- **Dur√©e totale des donn√©es saines (Classe_00)** : {:.3f} secondes ({} √©chantillons)
- **Dur√©e totale des donn√©es Outer Race (Classe_01)** : {:.3f} secondes ({} √©chantillons)

""".format(N_sains * delta_t, N_sains, N_outer_race * delta_t, N_outer_race)))

# --- 1. Signaux temporels (T) ---
fig_t = go.Figure()

# Ajouter les signaux des sujets sains (Classe_00)
fig_t.add_trace(go.Scatter(
    x=time_sains,
    y=df_sains["acceleration_vertbleu"],
    mode="lines",
    name="Sains - Acc√©l√©ration Vertbleu",
    line=dict(color="blue", width=1)
))
fig_t.add_trace(go.Scatter(
    x=time_sains,
    y=df_sains["acceleration_vert"],
    mode="lines",
    name="Sains - Acc√©l√©ration Vert",
    line=dict(color="cyan", width=1)
))

# Ajouter les signaux Outer Race (Classe_01)
fig_t.add_trace(go.Scatter(
    x=time_outer_race,
    y=df_outer_race["acceleration_vertbleu"],
    mode="lines",
    name="Outer Race - Acc√©l√©ration Vertbleu",
    line=dict(color="red", width=1)
))
fig_t.add_trace(go.Scatter(
    x=time_outer_race,
    y=df_outer_race["acceleration_vert"],
    mode="lines",
    name="Outer Race - Acc√©l√©ration Vert",
    line=dict(color="orange", width=1)
))

# Personnaliser le graphique
fig_t.update_layout(
    title=dict(
        text="Comparaison des signaux temporels (Fs = 25600 Hz)",
        x=0.5,
        font=dict(size=20)
    ),
    xaxis_title="Temps (s)",
    yaxis_title="Amplitude",
    legend_title="Signaux",
    template="plotly_white",
    width=1000,
    height=600,
    font=dict(size=14),
    margin=dict(l=50, r=50, t=80, b=50),
    showlegend=True
)

# Afficher le graphique temporel
display(Markdown("""
## Comparaison des signaux temporels üìà

Voici le trac√© des signaux temporels pour les deux fichiers (sains et Outer Race), avec une fr√©quence d'√©chantillonnage de 25600 Hz. Le graphique s'ouvrira dans votre navigateur par d√©faut.

"""))
fig_t.show()

# --- 2. FFT des signaux ---
# Calculer la FFT pour les signaux des sujets sains (Classe_00)
fft_sains_vertbleu = fft(df_sains["acceleration_vertbleu"].values)
fft_sains_vert = fft(df_sains["acceleration_vert"].values)

# Calculer la FFT pour les signaux Outer Race (Classe_01)
fft_outer_race_vertbleu = fft(df_outer_race["acceleration_vertbleu"].values)
fft_outer_race_vert = fft(df_outer_race["acceleration_vert"].values)

# Prendre la magnitude (valeur absolue) de la FFT
magnitude_sains_vertbleu = np.abs(fft_sains_vertbleu)
magnitude_sains_vert = np.abs(fft_sains_vert)
magnitude_outer_race_vertbleu = np.abs(fft_outer_race_vertbleu)
magnitude_outer_race_vert = np.abs(fft_outer_race_vert)

# Cr√©er le graphique Plotly pour la FFT
fig_fft = go.Figure()

# Ajouter les spectres des sujets sains (Classe_00)
fig_fft.add_trace(go.Scatter(
    x=frequencies_sains[:N_sains//2],
    y=magnitude_sains_vertbleu[:N_sains//2],
    mode="lines",
    name="Sains - Acc√©l√©ration Vertbleu",
    line=dict(color="blue", width=1)
))
fig_fft.add_trace(go.Scatter(
    x=frequencies_sains[:N_sains//2],
    y=magnitude_sains_vert[:N_sains//2],
    mode="lines",
    name="Sains - Acc√©l√©ration Vert",
    line=dict(color="cyan", width=1)
))

# Ajouter les spectres Outer Race (Classe_01)
fig_fft.add_trace(go.Scatter(
    x=frequencies_outer_race[:N_outer_race//2],
    y=magnitude_outer_race_vertbleu[:N_outer_race//2],
    mode="lines",
    name="Outer Race - Acc√©l√©ration Vertbleu",
    line=dict(color="red", width=1)
))
fig_fft.add_trace(go.Scatter(
    x=frequencies_outer_race[:N_outer_race//2],
    y=magnitude_outer_race_vert[:N_outer_race//2],
    mode="lines",
    name="Outer Race - Acc√©l√©ration Vert",
    line=dict(color="orange", width=1)
))

# Personnaliser le graphique FFT
fig_fft.update_layout(
    title=dict(
        text="Spectre fr√©quentiel des signaux (Fs = 25600 Hz)",
        x=0.5,
        font=dict(size=20)
    ),
    xaxis_title="Fr√©quence (Hz)",
    yaxis_title="Amplitude",
    legend_title="Spectres",
    template="plotly_white",
    width=1000,
    height=600,
    font=dict(size=14),
    margin=dict(l=50, r=50, t=80, b=50),
    showlegend=True,
    xaxis=dict(range=[0, 12800])  # Limiter √† la fr√©quence de Nyquist (12800 Hz)
)

# Afficher le graphique FFT
display(Markdown("""
## Spectre fr√©quentiel des signaux üîä

Voici le spectre fr√©quentiel des 4 signaux (sains et Outer Race), calcul√© avec une fr√©quence d'√©chantillonnage de 25600 Hz. La fr√©quence maximale affich√©e est la fr√©quence de Nyquist (12800 Hz) pour √©viter l'aliasing. Le graphique s'ouvrira dans votre navigateur par d√©faut.

"""))
fig_fft.show()

# --- 3. Enveloppe de Hilbert ---
# Calculer l'enveloppe de Hilbert pour les signaux des sujets sains (Classe_00)
envelope_sains_vertbleu = np.abs(hilbert(df_sains["acceleration_vertbleu"].values))
envelope_sains_vert = np.abs(hilbert(df_sains["acceleration_vert"].values))

# Calculer l'enveloppe de Hilbert pour les signaux Outer Race (Classe_01)
envelope_outer_race_vertbleu = np.abs(hilbert(df_outer_race["acceleration_vertbleu"].values))
envelope_outer_race_vert = np.abs(hilbert(df_outer_race["acceleration_vert"].values))

# Cr√©er le graphique Plotly pour l'enveloppe
fig_envelope = go.Figure()

# Ajouter les enveloppes des sujets sains (Classe_00)
fig_envelope.add_trace(go.Scatter(
    x=time_sains,
    y=envelope_sains_vertbleu,
    mode="lines",
    name="Sains - Enveloppe Vertbleu",
    line=dict(color="blue", width=1)
))
fig_envelope.add_trace(go.Scatter(
    x=time_sains,
    y=envelope_sains_vert,
    mode="lines",
    name="Sains - Enveloppe Vert",
    line=dict(color="cyan", width=1)
))

# Ajouter les enveloppes Outer Race (Classe_01)
fig_envelope.add_trace(go.Scatter(
    x=time_outer_race,
    y=envelope_outer_race_vertbleu,
    mode="lines",
    name="Outer Race - Enveloppe Vertbleu",
    line=dict(color="red", width=1)
))
fig_envelope.add_trace(go.Scatter(
    x=time_outer_race,
    y=envelope_outer_race_vert,
    mode="lines",
    name="Outer Race - Enveloppe Vert",
    line=dict(color="orange", width=1)
))

# Personnaliser le graphique de l'enveloppe
fig_envelope.update_layout(
    title=dict(
        text="Enveloppe de Hilbert des signaux (Fs = 25600 Hz)",
        x=0.5,
        font=dict(size=20)
    ),
    xaxis_title="Temps (s)",
    yaxis_title="Amplitude de l'enveloppe",
    legend_title="Enveloppes",
    template="plotly_white",
    width=1000,
    height=600,
    font=dict(size=14),
    margin=dict(l=50, r=50, t=80, b=50),
    showlegend=True
)

# Afficher le graphique de l'enveloppe
display(Markdown("""
## Enveloppe de Hilbert des signaux üìâ

Voici l'enveloppe de Hilbert des 4 signaux (sains et Outer Race), calcul√©e avec une fr√©quence d'√©chantillonnage de 25600 Hz. Le graphique s'ouvrira dans votre navigateur par d√©faut.

"""))
fig_envelope.show()

# --- 4. FFT de l'enveloppe ---
# Calculer la FFT des enveloppes
fft_envelope_sains_vertbleu = fft(envelope_sains_vertbleu)
fft_envelope_sains_vert = fft(envelope_sains_vert)
fft_envelope_outer_race_vertbleu = fft(envelope_outer_race_vertbleu)
fft_envelope_outer_race_vert = fft(envelope_outer_race_vert)

# Prendre la magnitude de la FFT des enveloppes
magnitude_envelope_sains_vertbleu = np.abs(fft_envelope_sains_vertbleu)
magnitude_envelope_sains_vert = np.abs(fft_envelope_sains_vert)
magnitude_envelope_outer_race_vertbleu = np.abs(fft_envelope_outer_race_vertbleu)
magnitude_envelope_outer_race_vert = np.abs(fft_envelope_outer_race_vert)

# Cr√©er le graphique Plotly pour la FFT de l'enveloppe
fig_fft_envelope = go.Figure()

# Ajouter les spectres des enveloppes des sujets sains (Classe_00)
fig_fft_envelope.add_trace(go.Scatter(
    x=frequencies_sains[:N_sains//2],
    y=magnitude_envelope_sains_vertbleu[:N_sains//2],
    mode="lines",
    name="Sains - Enveloppe Vertbleu",
    line=dict(color="blue", width=1)
))
fig_fft_envelope.add_trace(go.Scatter(
    x=frequencies_sains[:N_sains//2],
    y=magnitude_envelope_sains_vert[:N_sains//2],
    mode="lines",
    name="Sains - Enveloppe Vert",
    line=dict(color="cyan", width=1)
))

# Ajouter les spectres des enveloppes Outer Race (Classe_01)
fig_fft_envelope.add_trace(go.Scatter(
    x=frequencies_outer_race[:N_outer_race//2],
    y=magnitude_envelope_outer_race_vertbleu[:N_outer_race//2],
    mode="lines",
    name="Outer Race - Enveloppe Vertbleu",
    line=dict(color="red", width=1)
))
fig_fft_envelope.add_trace(go.Scatter(
    x=frequencies_outer_race[:N_outer_race//2],
    y=magnitude_envelope_outer_race_vert[:N_outer_race//2],
    mode="lines",
    name="Outer Race - Enveloppe Vert",
    line=dict(color="orange", width=1)
))

# Personnaliser le graphique FFT de l'enveloppe
fig_fft_envelope.update_layout(
    title=dict(
        text="Spectre fr√©quentiel des enveloppes (Fs = 25600 Hz)",
        x=0.5,
        font=dict(size=20)
    ),
    xaxis_title="Fr√©quence (Hz)",
    yaxis_title="Amplitude",
    legend_title="Spectres des enveloppes",
    template="plotly_white",
    width=1000,
    height=600,
    font=dict(size=14),
    margin=dict(l=50, r=50, t=80, b=50),
    showlegend=True,
    xaxis=dict(range=[0, 12800])  # Limiter √† la fr√©quence de Nyquist (12800 Hz)
)

# Afficher le graphique FFT de l'enveloppe
display(Markdown("""
## Spectre fr√©quentiel des enveloppes üîä

Voici le spectre fr√©quentiel des enveloppes de Hilbert des 4 signaux (sains et Outer Race), calcul√© avec une fr√©quence d'√©chantillonnage de 25600 Hz. La fr√©quence maximale affich√©e est la fr√©quence de Nyquist (12800 Hz). Le graphique s'ouvrira dans votre navigateur par d√©faut.

"""))
fig_fft_envelope.show()
```

```python
import os
import pandas as pd
import numpy as np
from scipy.fft import fft, fftfreq
from scipy.signal import hilbert
from scipy.stats import kurtosis, skew
from IPython.display import display, Markdown

# Chemins vers les dossiers
base_path = "C:/Users/bonij/OneDrive/Documents/cours ECC/2AA/S8/data driven decision/Base_de_donn√©es_S√©ance1"
sains_dir = f"{base_path}/Classe_00"
outer_race_dir = f"{base_path}/Classe_01"

# Fr√©quence d'√©chantillonnage
fs = 25600  # Hz

# Fonction pour extraire les features d'un signal
def extract_features(signal, fs):
    # Domaine temporel
    mean_val = np.mean(signal)
    std_val = np.std(signal)
    max_val = np.max(signal)
    min_val = np.min(signal)
    kurtosis_val = kurtosis(signal)
    skew_val = skew(signal)

    # Domaine fr√©quentiel (FFT)
    N = len(signal)
    fft_vals = fft(signal)
    frequencies = fftfreq(N, 1/fs)
    magnitude = np.abs(fft_vals)[:N//2]  # Partie positive du spectre
    freqs = frequencies[:N//2]
    
    dominant_idx = np.argmax(magnitude)
    dominant_freq = freqs[dominant_idx]
    dominant_amplitude = magnitude[dominant_idx]
    spectral_energy = np.sum(magnitude**2)

    # Enveloppe de Hilbert
    envelope = np.abs(hilbert(signal))
    envelope_mean = np.mean(envelope)
    envelope_std = np.std(envelope)

    # FFT de l'enveloppe
    fft_envelope = fft(envelope)
    magnitude_envelope = np.abs(fft_envelope)[:N//2]
    envelope_dominant_idx = np.argmax(magnitude_envelope)
    envelope_dominant_freq = freqs[envelope_dominant_idx]
    envelope_dominant_amplitude = magnitude_envelope[envelope_dominant_idx]

    return [
        mean_val, std_val, max_val, min_val, kurtosis_val, skew_val,
        dominant_freq, dominant_amplitude, spectral_energy,
        envelope_mean, envelope_std, envelope_dominant_freq, envelope_dominant_amplitude
    ]

# Liste des features
feature_names = [
    "Mean", "Std", "Max", "Min", "Kurtosis", "Skew",
    "Dominant Frequency", "Dominant Amplitude", "Spectral Energy",
    "Envelope Mean", "Envelope Std", "Envelope Dominant Frequency", "Envelope Dominant Amplitude"
]

# Initialiser les listes pour stocker les features et les labels
all_features = []
labels = []
signal_names = []  # Pour garder une trace des noms des signaux

# --- Traiter les fichiers sains (Classe_00) ---
sains_files = [f for f in os.listdir(sains_dir) if f.endswith(".xlsx")]
display(Markdown(f"### Nombre de fichiers sains (Classe_00) : {len(sains_files)}"))

for file in sains_files:
    file_path = os.path.join(sains_dir, file)
    try:
        df = pd.read_excel(file_path, engine="calamine")
        # Extraire les features pour acceleration_vertbleu
        features_vertbleu = extract_features(df["acceleration_vertbleu"].values, fs)
        all_features.append(features_vertbleu)
        labels.append(0)  # Label 0 pour sain
        signal_names.append(f"{file}_vertbleu")
        # Extraire les features pour acceleration_vert
        features_vert = extract_features(df["acceleration_vert"].values, fs)
        all_features.append(features_vert)
        labels.append(0)  # Label 0 pour sain
        signal_names.append(f"{file}_vert")
    except Exception as e:
        display(Markdown(f"Erreur lors du traitement du fichier {file} : {str(e)}"))

# --- Traiter les fichiers Outer Race (Classe_01) ---
outer_race_files = [f for f in os.listdir(outer_race_dir) if f.endswith(".xlsx")]
display(Markdown(f"### Nombre de fichiers Outer Race (Classe_01) : {len(outer_race_files)}"))

for file in outer_race_files:
    file_path = os.path.join(outer_race_dir, file)
    try:
        df = pd.read_excel(file_path, engine="calamine")
        # Extraire les features pour acceleration_vertbleu
        features_vertbleu = extract_features(df["acceleration_vertbleu"].values, fs)
        all_features.append(features_vertbleu)
        labels.append(1)  # Label 1 pour d√©fectueux
        signal_names.append(f"{file}_vertbleu")
        # Extraire les features pour acceleration_vert
        features_vert = extract_features(df["acceleration_vert"].values, fs)
        all_features.append(features_vert)
        labels.append(1)  # Label 1 pour d√©fectueux
        signal_names.append(f"{file}_vert")
    except Exception as e:
        display(Markdown(f"Erreur lors du traitement du fichier {file} : {str(e)}"))

# Convertir les features en une matrice numpy
feature_matrix = np.array(all_features).T  # Transpos√©e pour avoir (nombre de features, nombre de signaux)

# Afficher les dimensions de la matrice
total_signals = 2 * (len(sains_files) + len(outer_race_files))
display(Markdown(f"""
## Dimensions de la matrice d'information üìè

- **Nombre de features** : {feature_matrix.shape[0]}
- **Nombre total de signaux** : {feature_matrix.shape[1]} (2 signaux par fichier √ó ({len(sains_files)} fichiers sains + {len(outer_race_files)} fichiers Outer Race) = {total_signals})
"""))

# Cr√©er un DataFrame pour une visualisation plus claire
df_features = pd.DataFrame(feature_matrix.T, columns=feature_names)
df_features["Label"] = labels  # Ajouter la colonne des labels
df_features["Signal Name"] = signal_names  # Ajouter une colonne pour les noms des signaux

# Afficher un aper√ßu du dataset
display(Markdown("""
## Aper√ßu du dataset üóÇÔ∏è

Voici les 5 premi√®res lignes du dataset (chaque ligne repr√©sente un signal, chaque colonne une feature) :
"""))
display(df_features.head())

# Sauvegarder le dataset dans un fichier CSV
df_features.to_csv("signal_features_dataset.csv", index=False)
display(Markdown("""
## Sauvegarde du dataset üíæ

Le dataset a √©t√© sauvegard√© dans le fichier `signal_features_dataset.csv` dans le r√©pertoire de travail de votre notebook.
"""))
```
